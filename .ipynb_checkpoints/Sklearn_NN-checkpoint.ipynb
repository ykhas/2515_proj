```python
from project.gen_data import gen_data
from project.post_processing import Plotter
```

```python
import matplotlib.pyplot as plt

data_dict = {
    "x_range": (0, 1),
    "t_range": (0, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": 0.4,
    "frequency": 1,
}

X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
Plotter(1,1).plot_2d_y(X_gen, y_gen)
plt.show()
```

```python
def draw_2d_graph(X_gen,y_gen):
    plt.clf()
    plt.xlabel("X")
    plt.ylabel("t")
    plt.pcolormesh(y_gen.reshape(100,100))
    plt.colorbar()
    return plt

draw_2d_graph(X_gen,y_gen)
print(y_gen.shape)
```

## Naive MLPRegressor of Default Single Hidden Layer 
This block is majorly a naive implementation of single layer Neural Network with sklearn, which derive a OK model with 98.7% accuracy. As showed below

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

y_gen_array_for_sklearn = y_gen.ravel()
X_train, X_test, y_train, y_test = train_test_split(X_gen, y_gen_array_for_sklearn)

regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
print("Test Score: ", regr.score(X_test, y_test))
y_pred = regr.predict(X_gen)
Plotter(1,1).plot_2d_y(X_gen, y_pred)
plt.title("3d graph of predicted model using single layer MLPRegression")
plt.show()
plt = draw_2d_graph(X_gen,y_pred)
plt.title("2d graph of predicted model using single layer MLPRegression")
plt.show()
y_dif = y_pred.reshape(y_gen.shape) - y_gen
Plotter(1,1).plot_2d_y(X_gen, y_dif)
plt.title("3d graph of errors from model to numerical data")
plt.show()
plt = draw_2d_graph(X_gen,y_dif)
plt.title("2d graph of errors from model to numerical data")
plt.show()
#draw_2d_graph(X_test, y_pred)
```

## Grid Search for MLPRegressor to Utilize the hyper-parameters
What we do here is try to train a better model that have more complex structures. We will use some of the tools that are partially used from previous homework, and turned out to be that a better model can be derived from the case

In case we lost the training results, the best methods we found is:

{'activation': 'relu', 'hidden_layer_sizes': (100, 150, 150)}

Test Score: 0.989

```python
# For hidden layer combinations, we use https://stackoverflow.com/questions/3099987/generating-permutations-with-repetitions as a refererence
import itertools
Activations = ['logistic','tanh', 'relu']
Hidden_Layers = []
Node_options = [60,100,150]
for i in range(1,4):
    layer_options = [p for p in itertools.product(Node_options, repeat=i)]
    Hidden_Layers.extend(layer_options)
print(Hidden_Layers)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import metrics

param_grid = {
        'hidden_layer_sizes': Hidden_Layers,
        'activation': Activations
    }
regr_grid = GridSearchCV(estimator=MLPRegressor(), param_grid=param_grid, verbose=3 )
regr_grid.fit(X_train, y_train)
```

```python
print("The best score over Grid Search", regr.best_score_)
```

```python
regr_grid.best_params_
```

```python
regr_best = regr_grid.best_estimator_
```

```python
y_pred = regr_best.predict(X_gen)
Plotter(1,1).plot_2d_y(X_gen, y_pred)
plt.title("3d graph of predicted model using single layer MLPRegression")
plt.show()
plt = draw_2d_graph(X_gen,y_pred)
plt.title("2d graph of predicted model using single layer MLPRegression")
plt.show()
y_dif = y_pred.reshape(y_gen.shape) - y_gen
Plotter(1,1).plot_2d_y(X_gen, y_dif)
plt.title("3d graph of errors from model to numerical data")
plt.show()
plt = draw_2d_graph(X_gen,y_dif)
plt.title("2d graph of errors from model to numerical data")
plt.show()
```

## Deeper Networks for MLPRegressor to Utilize the hyper-parameters
What we do here is try to train a better model that have more complex structures. We will use some of the tools that are partially used from previous homework, and turned out to be that a better model can be derived from the case

In case we lost the training results, the best methods we found is:

{'activation': 'relu', 'hidden_layer_sizes': (150, 200, 150, 150)}

Test Score: 0.0.999805139215787

```python
Activations = ["relu"]
Hidden_Layers = []
Node_options = [60,100,150,200]
for i in range(3,5):
    layer_options = [p for p in itertools.product(Node_options, repeat=i)]
    Hidden_Layers.extend(layer_options)
print(len(Hidden_Layers))
```

```python
param_grid = {
        'hidden_layer_sizes': Hidden_Layers,
        'activation': Activations
    }
regr_grid_more_layers = GridSearchCV(estimator=MLPRegressor(), param_grid=param_grid, verbose=3 )
regr_grid_more_layers.fit(X_train, y_train)
```

```python
regr_grid_more_layers.best_params_
```

```python
regr_grid_more_layers.best_score_
```

```python
regr_more_best = regr_grid_more_layers.best_estimator_
```

```python
from sklearn import metrics
metrics.explained_variance_score(y_test, regr_more_best.predict(X_test))
```

```python
y_pred = regr_more_best.predict(X_gen)
Plotter(1,1).plot_2d_y(X_gen, y_pred)
plt.title("3d graph of predicted model using single layer MLPRegression")
plt.show()
plt = draw_2d_graph(X_gen,y_pred)
plt.title("2d graph of predicted model using single layer MLPRegression")
plt.show()
y_dif = y_pred.reshape(y_gen.shape) - y_gen
Plotter(1,1).plot_2d_y(X_gen, y_dif)
plt.title("3d graph of errors from model to numerical data")
plt.show()
plt = draw_2d_graph(X_gen,y_dif)
plt.title("2d graph of errors from model to numerical data")
plt.show()
```
