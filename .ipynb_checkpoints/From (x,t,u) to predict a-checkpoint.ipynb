```python
from project.gen_data import gen_data
from project.post_processing import Plotter
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

data_dict = {
    "x_range": (0, 1),
    "t_range": (0, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": 0.4,
    "frequency": 1,
}

X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
Plotter(1,1).plot_2d_y(X_gen, y_gen)
plt.show()
```

```python
a_gen = np.linspace(0.01, 1.0, 20)
print(a_gen)
```

```python
X_all=np.array([])
y_all=[]
for a in a_gen:
    data_dict = {
    "x_range": (0, 1),
    "t_range": (0, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": a,
    "frequency": 1,
    }
    X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
    X_u = np.concatenate((X_gen, y_gen), axis=1)
    if (X_all.shape[0] < 2):
        X_all = X_u
    else:
        X_all = np.concatenate((X_all, X_u), axis=0)
    for i in range(X_u.shape[0]):
        y_all.append(a)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all)
```

```python
regr = MLPRegressor(hidden_layer_sizes = (150, 200, 150, 150,200),activation='relu',tol=1e-6,verbose=1).fit(X_train, y_train)
```

```python
metrics.explained_variance_score(y_test, regr.predict(X_test))
```

```python
index_50 = [np.random.randint(0,X_all.shape[0]) for i in range(50)]
X_rand = X_all[index_50]
a_pred = regr.predict(X_rand)
for i in range(50):
    a_true = y_all[index_50[i]]
    ratio = (a_pred[i] - a_true)/a_true
    print("Random Check. Truth: {:.2f}, Predict: {:.2f}, ratio: {:.2f}".format(y_all[index_50[i]], a_pred[i], ratio))
```

```python
regr_smaller = MLPRegressor(hidden_layer_sizes = (150, 200, 150),activation='relu',tol=1e-6,verbose=3).fit(X_train, y_train)
metrics.explained_variance_score(y_test, regr_smaller.predict(X_test))
```

```python
index_50 = [np.random.randint(0,X_all.shape[0]) for i in range(50)]
X_rand = X_all[index_50]
a_pred = regr_smaller.predict(X_rand)
for i in range(50):
    a_true = y_all[index_50[i]]
    ratio = (a_pred[i] - a_true)/a_true
    print("Random Check. Truth: {:.2f}, Predict: {:.2f}".format(y_all[index_50[i]], a_pred[i]))
```

```python
for a in [0.1, 0.3, 0.5, 0.7, 0.9]:
    data_dict = {
    "x_range": (0, 1),
    "t_range": (0, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": a,
    "frequency": 1,
    }
    X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
    X_u = np.concatenate((X_gen, y_gen), axis=1)
    a_pred = regr_smaller.predict(X_u)
    print("for a = {:.1f}, Average prediction: {:.3f}, STD: {:.3f}".format(a, np.average(a_pred), np.std(a_pred)))
    a_error = a_pred - a
    plt.clf()
    plt.xlabel("X")
    plt.ylabel("t")
    plt.pcolormesh(y_gen.reshape(100,100))
    plt.colorbar()
    plt.title("Error distribution over X and t, a = {}".format(a))
    plt.show()
```

## Utilization

Since for t = 0, all the data are exactly the same, which is not something can determine anything in general. So we will delete all the data that t < 0.02, to see if things go differently

```python
a_gen = np.linspace(0.01, 1.0, 20)
X_all=np.array([])
y_all=[]
for a in a_gen:
    data_dict = {
    "x_range": (0, 1),
    "t_range": (0.02, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": a,
    "frequency": 1,
    }
    X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
    X_u = np.concatenate((X_gen, y_gen), axis=1)
    if (X_all.shape[0] < 2):
        X_all = X_u
    else:
        X_all = np.concatenate((X_all, X_u), axis=0)
    for i in range(X_u.shape[0]):
        y_all.append(a)
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all)
regr_t_filter = MLPRegressor(hidden_layer_sizes = (150, 200, 150, 150),activation='relu',tol=1e-6,verbose=3).fit(X_train, y_train)
metrics.explained_variance_score(y_test, regr_t_filter.predict(X_test))
```

```python
for a in [0.1, 0.3, 0.5, 0.7, 0.9]:
    data_dict = {
    "x_range": (0, 1),
    "t_range": (0.02, 1),
    "x_dim": 100,
    "t_dim": 100,
    "a_coeff": a,
    "frequency": 1,
    }
    X_gen, y_gen = gen_data("heat_1d_boundary_sin_exact", data_dict)
    X_u = np.concatenate((X_gen, y_gen), axis=1)
    a_pred = regr_smaller.predict(X_u)
    print("for a = {:.1f}, Average prediction: {:.3f}, STD: {:.3f}".format(a, np.average(a_pred), np.std(a_pred)))
    a_error = a_pred - a
    plt.clf()
    plt.xlabel("X")
    plt.ylabel("t")
    plt.pcolormesh(y_gen.reshape(100,100))
    plt.colorbar()
    plt.title("Error distribution over X and t, a = {}".format(a))
    plt.show()
```

Not too much difference, means parameter guessing over traditional NN is not a good idea in general
